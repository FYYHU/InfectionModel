```python
INFECTED_P = 0.05
INFECTION_P = 0.1
RECOVERY_T = 14
RECOVERY_W = 5

G = nx.erdos_renyi_graph(500, 0.01)
FREQ_DIST = 'uniform'

N_DAYS = 100
N_SIMS = 100
SHARE_INIT = True
ANALYSIS_DAY = 4

edge_gen = edge_gen_from_nx(G, N_DAYS, freq_dist=FREQ_DIST)
model = SimModel(G, edge_gen, INFECTED_P, INFECTION_P, RECOVERY_T, RECOVERY_W, N_DAYS, tqdm=tqdm)
model.run_sim()
```
